###Tabular Q-learning

```
seed = 1234
set_seed(seed)

decay = 1000
policy = EpsilonGreedy(decay=decay)

lr = .1
method = 'tabular'
degree = 2
agent = QLearning(policy=policy,lr=lr,method=method,degree=degree)

env = Environment(sale=3,wholesale=5,retail=7,r=.1,gamma=1/(1+.1),N=7,M=20,lam=9)

episodes = 20000
agent.fit(env=env,episodes=episodes,verbose_freq=.1,plot_freq=[episodes/2,episodes])
```
